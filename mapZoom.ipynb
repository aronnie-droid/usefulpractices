```python
import arcpy

def _merkez():
    aprx = arcpy.mp.ArcGISProject("CURRENT")
    mv = aprx.activeView

    # aktif view kontrolü
    if mv is None or not hasattr(mv, "camera"):
        raise Exception("Aktif bir harita görünümü yok. Lütfen bir Map View seçin.")

    ext = mv.camera.getExtent()
    sr = mv.map.spatialReference
    
    pt = arcpy.PointGeometry(
        arcpy.Point((ext.XMin+ext.XMax)/2, (ext.YMin+ext.YMax)/2),
        sr
    ).projectAs(arcpy.SpatialReference(4326))
    
    return pt.centroid.X, pt.centroid.Y

def k():
    try:
        lon, lat = _merkez()
        print(f"{lon:.5f},{lat:.5f},18")
    except Exception as e:
        print(e)

def h():
    try:
        lon, lat = _merkez()
        print(f"https://harita.istanbul/2d?@={lon:.5f},{lat:.5f},18&ms=!b241!c")
    except Exception as e:
        print(e)

def h3b():
    try:
        lon, lat = _merkez()
        print(f"https://harita.istanbul/3d?@={lon:.5f},{lat:.5f},18")
    except Exception as e:
        print(e)

def g():
    try:
        lon, lat = _merkez()
        print(f"https://www.google.com/maps?q={lat:.5f},{lon:.5f}")
    except Exception as e:
        print(e)

def y():
    try:
        lon, lat = _merkez()
        print(f"https://yandex.com.tr/maps/?ll={lon:.5f}%2C{lat:.5f}&z=17&l=sat")
    except Exception as e:
        print(e)

def ea():
    try:
        lon, lat = _merkez()
        print(f"https://earth.google.com/web/@{lat:.5f},{lon:.5f},1000a,17d,35y,0h,0t,0r")
    except Exception as e:
        print(e)

def m():
    try:
        lon, lat = _merkez()
        print(f"https://www.mapillary.com/app/?lat={lat:.5f}&lng={lon:.5f}&z=17")
    except Exception as e:
        print(e)

def w():
    try:
        lon, lat = _merkez()
        print(f"http://wikimapia.org/#lat={lat:.5f}&lon={lon:.5f}&z=17")
    except Exception as e:
        print(e)
```

```python
k()  # 29.21330,41.00890,17
h()  # harita.istanbul link
h3b()# harita.istanbul 3d link
g()  # Google Maps link
ea() # Google earth link
y()  # Yandex Maps link
m()  # Mapillary link
w()  # Wikimapia link
```
