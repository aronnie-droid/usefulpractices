```python
# ===============================
# HÜCRE 0 — GEREKLİ KÜTÜPHANELER
# ===============================

import arcpy
import pandas as pd
import geopandas as gpd
import numpy as np
import requests
import os
import json

from collections import defaultdict
from arcgis.gis import GIS
from arcgis.features import FeatureLayer
from arcgis.features import FeatureSet
```

```python
# ==========================================
# HÜCRE 1 — PORTAL BAĞLANTISI VE VERİLER
# ==========================================

# ---- 1. Portala Bağlan ----
portal_url = "https://geo.ipa.istanbul/portal"
gis = GIS(portal_url, "kullanici.adi", "kullanici.sifresi")

print("Portal bağlantısı başarılı:", gis.properties.portalHostname)

# ---- 2. İlçe Layer → GeoDataFrame ----
ilce_url = "https://geo.ipa.istanbul/server/rest/services/Hosted/ilce_turef30/FeatureServer/0"
ilce_layer = FeatureLayer(ilce_url)

ilce_fs = ilce_layer.query(where="1=1", out_fields="*", return_geometry=True)
ilce_gdf = gpd.read_file(ilce_fs.to_geojson)

print("İlçe GDF yüklendi. Kayıt sayısı:", len(ilce_gdf))

# ---- 3. Mahalle Layer → GeoDataFrame ----
mahalle_url = "https://geo.ipa.istanbul/server/rest/services/Hosted/mahalle_turef30/FeatureServer/0"
mahalle_layer = FeatureLayer(mahalle_url)

mahalle_fs = mahalle_layer.query(where="1=1", out_fields="*", return_geometry=True)
mahalle_gdf = gpd.read_file(mahalle_fs.to_geojson)

print("Mahalle GDF yüklendi. Kayıt sayısı:", len(mahalle_gdf))

# ---- 4. Detaylı AK Layer → GeoDataFrame ----
detayli_ak_url = "https://geo.ipa.istanbul/server/rest/services/Hosted/detayli_ak/FeatureServer/0"
detayli_ak_layer = FeatureLayer(detayli_ak_url)

detayli_ak_fs = detayli_ak_layer.query(where="1=1", out_fields="*", return_geometry=True)
detayli_ak_gdf = gpd.read_file(detayli_ak_fs.to_geojson)

print("Detaylı AK GDF yüklendi. Kayıt sayısı:", len(detayli_ak_gdf))

# ---- 5. Alt Kullanım Açıklama CSV ----
csv_path = r"C:\Users\___.korucu\Desktop\veri_temin_____\scripts\domains.csv"
alt_kullanim_df = pd.read_csv(csv_path)

print("Alt kullanım açıklama tablosu yüklendi. Kayıt sayısı:", len(alt_kullanim_df))


# ---- 6. Kırsal Mahalleler Excel ----
excel_path = r"excel_yolu"
kirsal_df = pd.read_excel(excel_path)

print("Kırsal mahalleler tablosu yüklendi. Kayıt sayısı:", len(kirsal_df))
```
